```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gn = np.random.default_rng(seed=42)
rx = gn.normal(loc=0, scale=1)
ry = gn.normal(loc=0, scale=1)
def first(x , y):
    return (x - 2)**2 + (y - 2)**2 + 5

def f_dev(x):
    return 2*(x - 2)

def s_dev(y):
    return 2*(y - 2)
```

```python
grad_x = [rx]
grad_y = [ry]
lr = 0.01
epochs = 100
for i in range(epochs):

    rx = rx - f_dev(rx) * lr
    ry = ry - s_dev(ry) * lr

    if (i % 10) == 0:

        grad_x.append(rx)
        grad_y.append(ry)
```

```python
#contour points
x_range = np.linspace(-1, 4, 100)
y_range = np.linspace(-1, 4, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = first(X, Y)

plt.figure(figsize=(8, 6))

plt.contour(X, Y, Z, levels=20, cmap='viridis')

plt.plot(grad_x, grad_y, 'ro-', label='GD Path') 

plt.plot(grad_x[0], grad_y[0], 'go', label='Start')
plt.plot(2, 2, 'bx', markersize=10, label='Minimum (2,2)')

plt.title("Gradient Descent Path on $f(x, y) = (x-2)^2 + (y-2)^2 + 5$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
