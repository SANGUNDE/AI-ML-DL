```python

from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creation of datasets
X, y = make_regression(n_samples=20, n_features=1, n_informative=1, random_state=42)
```

```python
def standardizing(df):
    global_std = df.std()
    global_mean = df.mean()
    return np.array([(data - global_mean) / global_std for data in df])
```

```python
scaled_X = standardizing(X)
```

```python
def data_split(X,y, amount):
    initial = int(len(X))
    test = int(len(X) * amount)
    train = int(initial - test)
    return np.array(X[:train]), np.array(X[train:]), np.array(y[:train]), np.array(y[train:])
```

```python
X_train, X_test, y_train, y_test = data_split(scaled_X, y, 0.2)
```

```python
print('shape train X', X_train.shape)
print('shape train y', y_train.shape)
print('shape test X', X_test.shape)
print('shape test y', y_test.shape)
print('shape full data', scaled_X.shape)
```

```python
def regression_fit(train_x, train_y, n_iter=40, learning_rate=0.1):
    rng = np.random.default_rng(seed=42)
    n_samples, n_features = train_x.shape
    train_y = train_y.ravel()


    weights = rng.normal(loc=0.0, scale=1.0, size=(n_features,))
    bias = 0

    for i in range(n_iter):

        y_pred = np.dot(train_x, weights) + bias


        error = y_pred - train_y
        loss = np.mean(error**2)

        dw = (2 / n_samples) * np.dot(train_x.T, error)
        db = (2 / n_samples) * np.sum(error)

        weights -= learning_rate * dw

        bias -= learning_rate * db

        if i % 2 == 0:
            print(f'epochs: {i} | loss: {loss}')

    return weights, bias
```

```python
weights, bias = regression_fit(X_train,y_train)
```

```python
def prediction(X, weights, bias):
    return np.dot(X, weights) + bias
```

```python
pd.DataFrame({
    'real data': y_test,
    'predicted': prediction(X_test, weights, bias)
})
```
